```python
import matplotlib.pyplot as plt
```

# Bar Plot


```python
x = [1,2,3,4,5,6]
y = [25,35,14,73,82,43]
plt.bar(x,y)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [25,35,14,73,82,43]
plt.xlabel("x-axis",fontsize = 10)
plt.ylabel("y-axis", fontsize = 20)
plt.title("Random Bar Plot")
plt.bar(x,y)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [25,35,14,73,82,43]
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Random Bar Plot")
plt.bar(x,y, width = 0.5, color = "r")
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [25,35,14,73,82,43]
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Random Bar Plot")
c = ["red","blue","yellow","green","purple","orange"]
plt.bar(x,y, width = 0.5, color = c,align = "edge")
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [25,35,14,73,82,43]
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Random Bar Plot")
c = ["red","blue","yellow","green","purple","orange"]
plt.bar(x,y, width = 0.5, color = c,align = "edge",edgecolor ="black", linewidth=5, linestyle = ":")
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [25,35,14,73,82,43]
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Random Bar Plot")
c = ["red","blue","yellow","green","purple","orange"]
plt.bar(x,y, width = 0.5, color = c,align = "edge",edgecolor ="black", linewidth=5, linestyle = ":",alpha = 0.5)
plt.show()
```
